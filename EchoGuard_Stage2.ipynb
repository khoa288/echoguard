```python
import json, glob, os

for d in sorted(glob.glob("runs_onefold/*")):
    if not os.path.isdir(d):
        continue
    cfg = os.path.join(d, "config.json")
    if os.path.isfile(cfg):
        with open(cfg, "r") as f:
            c = json.load(f)
        model = c.get("model") or c.get("models")
        tau = c.get("tau")
        print(os.path.basename(d), "model=", model, "tau=", tau)
```
